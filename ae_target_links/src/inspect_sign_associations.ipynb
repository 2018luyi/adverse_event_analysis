```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
target_names = pd.read_csv('/scratch/ias41/ae_code/ae_target_links/data/target_names.txt', sep='\t')
```

```python
def get_all_associations(output_dir):
    all_targets_df = pd.DataFrame()
    
    files = [i for i in os.listdir(output_dir + '/data') if 'ratios.txt' in i]
    for file in files:
        
        target_df = pd.read_csv(output_dir + '/data/{}'.format(file), sep='\t')
        target_df = target_df.loc[(target_df['nr compounds with AE']>=5)&(target_df['nr compounds active']>=5),:]
        
        if len(target_df) < 1:
            continue
        
        significant = target_df.loc[(target_df['corrected p-value']< 0.05)&(target_df['Likelihood Ratio']>= 2),:]
        if len(significant) > 0:
            all_targets_df = all_targets_df.append(significant, sort=False)
    
    return all_targets_df.merge(target_names, on='accession', how='left').sort_values(by='pref_name')
```

```python
all_aes = get_all_associations('/scratch/ias41/ae_code/ae_target_links/output/20191229_sider_unbound_margin_pred')
```
